```python
from ast import literal_eval
```

```python
movies["genres"]=movies["genres"].fillna("[]").apply(literal_eval).apply(lambda x:[i["name"] for i in x] if isinstance(x,list) else [] )
```

```python
movies.head()
```

```python
vote_counts=movies[movies["vote_count"].notnull()]["vote_count"].astype("int")
```

```python
vote_averages=movies[movies["vote_average"].notnull()]["vote_average"].astype("int")
```

```python
c=vote_averages.mean()
```

```python
c
```

```python
m=vote_counts.quantile(0.95)
```

```python
movies["year"] = pd.to_datetime(movies["release_date"], errors="coerce").apply(lambda x: str(x).split("-")[0] if pd.notnull(x) else np.nan)
```

```python
movies
```

```python
qualified=movies[(movies["vote_count"]>=m) & (movies["vote_count"].notnull()) & (movies["vote_average"].notnull())][["title","year","vote_count","vote_average","popularity","genres"]]
```

```python
qualified["vote_average"]=qualified["vote_average"].astype(int)
```

```python
qualified.info()
```

```python
qualified.shape
```

```python
def weighted_rating(x):
    v=x["vote_count"]
    r=x["vote_average"]
    
    return ((v/(v+m))*r + (m/m+v)*c)
```

```python
qualified["WR"]=qualified.apply(weighted_rating, axis=1)
```

```python
qualified
```

```python
qualified=qualified.sort_values("WR",ascending=False).head(250)
```
